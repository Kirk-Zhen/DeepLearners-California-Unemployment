```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
pd.set_option('expand_frame_repr', False)

# set Date as datetype datetime (for sorting purpose)
df['Date'] = pd.to_datetime(df['Date'])
# Remove seasonally adjusted data
N_df = df[(df["Seasonally Adjusted (Y/N) "] == "N")]
```

```python
rnn_df = pd.DataFrame(data=None, index=None, dtype=None, copy=False,
                      columns=["area_name", "labor_force", "employment", "unemployment", "unemployment_rate"])

areas = N_df["Area Name "].unique()
for i in range(len(areas)):
    area = areas[i]
    t_df = N_df[(df["Area Name "] == area)]
    t_df = t_df.sort_values(by='Date')
    labor = t_df["Labor Force"].tolist()
    employment = t_df["Employment "].tolist()
    unemployment = t_df["Unemployment "].tolist()
    u_rate = t_df["Unemployment Rate "].tolist()
    assert len(labor) == len(unemployment) and len(unemployment) == len(employment) and len(employment) == len(u_rate), f"some data missed for {area}"
    
    df_temp = {'area_name': area, 
           'labor_force': labor, 
           'employment': employment, 
           'unemployment': unemployment, 
           'unemployment_rate': u_rate}
    rnn_df = rnn_df.append(df_temp, ignore_index = True)
```

```python
rnn_df.to_pickle("rnn_dataframe.pkl")  
```
